### Coding from scratch

```python
import nltk
import numpy as np
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
#word_extraction : remove all stop word and return the filtered text
def word_extraction(text):
    stop_words = set(stopwords.words("english"))
    word_tokens = word_tokenize(text.lower())
    filtered_text = [word for word in word_tokens if word not in stop_words]
    return filtered_text

#tokenize_all : get all tokens and return unique values of them
def tokenize_all(texts):
    words = []
    for text in texts:
        words.extend(word_extraction(text))
    return sorted(list(set(words)))

#generated_bow : generated the Bag of Word of each text
def generate_bow(texts):
    bags = {}
    vocabs = tokenize_all(texts)
    for index, text in enumerate(texts):
        words = word_extraction(text)
        bag = np.zeros(len(vocabs))
        for word in words:
            for id, vocab in enumerate(vocabs):
                if vocab == word:
                    bag[id] += 1
        bags[index] = bag
    return bags

```

### BOW using sklearn

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

#define the corpus
corpus = ["i cant wait to get out of lockdown", 
          "the uk is soon going to be free soon",
          "linkedin is social media"]

#generate the BoW
bow = CountVectorizer()
bow.fit(corpus)
X = bow.transform(corpus)

#converting to dataframe
X_df = pd.DataFrame(X.toarray(), columns=sorted(bow.vocabulary_))
X_df
```
